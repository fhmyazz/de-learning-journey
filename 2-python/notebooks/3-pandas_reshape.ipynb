```python
# GP-1 — Buat pivot table yang menampilkan total revenue per month (baris) dan per category (kolom). Ganti NaN dengan 0.
# Hint: pivot_table() dengan index, columns, values, aggfunc, fill_value

completed_trx = transactions[transactions["status"] == "completed"].copy()
completed_trx["month"] = completed_trx["trx_date"].dt.strftime("%Y-%m")
pivoted_trx = completed_trx.pivot_table(
    index = "month",
    columns = "category",
    values = "amount",
    aggfunc = "sum",
    fill_value = 0
)

pivoted_trx

```

```python
# GP-2 — Dari DataFrame completed_trx, tambahkan tiga kolom:
# day_of_week — nama hari transaksi (Monday, Tuesday, dst)
# is_weekend — True kalau hari Sabtu atau Minggu, False kalau tidak
# days_ago — berapa hari yang lalu transaksi terjadi (dari hari ini)
# Hint: dt.day_name() untuk nama hari, dt.dayofweek untuk angka hari (0=Monday, 6=Sunday)

completed_trx["day_of_week"] = completed_trx["trx_date"].dt.day_name()

def weekend_check(day):
  if day == "Saturday" or day == "Sunday":
    return True
  else:
    return False

completed_trx["is_weekend"] = completed_trx["day_of_week"].apply(weekend_check)
#cara di bawah lebih pythonic
# completed_trx["is_weekend"] = completed_trx["trx_date"].dt.day_of_week >= 5

currDate = pd.Timestamp.now()
completed_trx["days_ago"] = (currDate - completed_trx["trx_date"]).dt.days
completed_trx
```
